## title.episode.tsv.gz

```python
# Variablen definieren um übergebene Werte von Pipeline-Parametern aufzunehmen
run_year = ""
run_month = ""
run_day = ""
```

```python
from pyspark.sql.types import *
from pyspark.sql.functions import col

#path = "Files/2024/01/16/"
path = "Files/" + run_year + "/" + run_month + "/" + run_day + "/"
```

```python
filepath = path + "title.episode.tsv.gz"
title_episode = spark.read \
    .options(    
        format="csv",
        header=True,
        delimiter="\t",
        compression="gzip",
        nullValue="\\N"
    ) \
    .csv(filepath)

display(title_episode.count())
title_episode.printSchema()
display(title_episode.limit(10))
```

```python
# Um Reaload der Textdatei zu umgehen sowie zwecks einfacherer Wiederverwendung von Code,
# wird Datenset einer neuen df-Variabel übergeben
df = title_episode
```

```python
# dient insbesondere der Überprüfung, ob Textwerte in numerischen Spalten vorkommen (Hinweis auf verrutschte Spalten)
df.summary("count", "mean", "min", "max").show()
```

```python
df = df.select(
    "tconst",
    "parentTconst",
    col("seasonNumber").cast("integer"),
    col("episodeNumber").cast("integer"))
df.printSchema()
```

```python
df.write.mode("overwrite").parquet(path + "title.episode.parquet")
df.write.mode("overwrite").format("delta").saveAsTable("silver_title_episode")
```
